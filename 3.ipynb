```python
from collections import deque

def solution(jug1_capacity, jug2_capacity, target):
    # Initialize visited states dictionary and result path
    visited = {}
    path = []
    
    # Create queue and add initial state (0,0)
    queue = deque([(0, 0)])
    is_solvable = False
    
    while queue:
        # Get current state
        current = queue.popleft()
        
        # Skip if already visited
        if current in visited:
            continue
            
        # Check if state is valid
        if (current[0] > jug1_capacity or 
            current[1] > jug2_capacity or 
            current[0] < 0 or current[1] < 0):
            continue
            
        # Add to path and mark as visited
        path.append([current[0], current[1]])
        visited[current] = 1
        
        # Check if target is reached
        if current[0] == target or current[1] == target:
            is_solvable = True
            if current[0] == target:
                if current[1] != 0:
                    path.append([current[0], 0])
            else:
                if current[0] != 0:
                    path.append([0, current[1]])
                    
            # Print solution path
            print("Solution Path:")
            for state in path:
                print(f"({state[0]}, {state[1]})")
            return
        
        # Generate next possible states
        # Fill jug 2
        queue.append((current[0], jug2_capacity))
        # Fill jug 1
        queue.append((jug1_capacity, current[1]))
        
        # Pour water between jugs
        for amount in range(max(jug1_capacity, jug2_capacity) + 1):
            # Pour from jug 2 to jug 1
            c = current[0] + amount
            d = current[1] - amount
            if c == jug1_capacity or (d == 0 and d >= 0):
                queue.append((c, d))
                
            # Pour from jug 1 to jug 2
            c = current[0] - amount
            d = current[1] + amount
            if (c == 0 and c >= 0) or d == jug2_capacity:
                queue.append((c, d))
        
        # Empty jugs
        queue.append((jug1_capacity, 0))
        queue.append((0, jug2_capacity))
    
    if not is_solvable:
        print("Solution not possible")

if __name__ == '__main__':
    jug1, jug2, target = 4, 3, 2
    print("Path from initial state to solution state:")
    solution(jug1, jug2, target)
```
